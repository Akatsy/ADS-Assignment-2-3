```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("WA_Fn-UseC_-HR-Employee-Attrition.csv")

```

```python
newdf=df.drop(["Age","StandardHours","StockOptionLevel","OverTime",'EmployeeCount',
       'EmployeeNumber','EnvironmentSatisfaction',"Over18","BusinessTravel","JobLevel",'YearsWithCurrManager',"DailyRate","MaritalStatus","NumCompaniesWorked","Gender","TotalWorkingYears","HourlyRate"],axis=1)
newdf.head()
```

```python
newdf.shape
```

```python
g=sns.countplot(x="Attrition",data=newdf)
g.set_title("Count of Attritions")
sns.set_style("whitegrid")
```

```python
a= sns.catplot(x="Department",data=newdf,kind="count",hue="Attrition")
a.fig.suptitle("Attritions per Department")
plt.xticks(rotation=90)
sns.set_style("darkgrid")
```

```python
m=sns.catplot(x="JobRole",y="DistanceFromHome",kind="bar",data=newdf,hue="Attrition",ci=None)
m.fig.suptitle("JobRole Vs Distance from Home",y=1.03)
sns.set_context("talk")
plt.xticks(rotation=90)
sns.set_style("dark")
```

```python
n=sns.catplot(x="Education",y="MonthlyIncome",hue="Attrition",data=newdf,kind="bar",ci=None)
n.fig.suptitle("Education Vs MonthlyIncome",y=1.03)
sns.set_context("talk")
sns.set_style("whitegrid")
plt.show()
```

```python
v=sns.relplot(x="JobSatisfaction",y="JobInvolvement",data=df,kind="line",hue="Attrition",col="Department",col_wrap=3,ci=None,dashes=False)
sns.set_style("ticks")
v.fig.suptitle("Attrition on JobInvolvement  vs JobSatisfaction per Department",y=1.03)
```
